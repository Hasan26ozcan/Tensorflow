```python
numpyDizisi1=np.linspace(0,10,100)
```

```python
numpyDizisi1
```

```python
plt.plot(numpyDizisi1)
```

```python
numpyDizisi2=numpyDizisi1**3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"g")
```

```python
plt.subplot(1,2,1) #1 sra olacak 2 kolon olacak uan 1. grafik iziliyor
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) #1 sra olacak 2 kolon olacak uan 2. grafik iziliyor
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure=plt.figure()
figureAxes=benimFigure.add_axes([0.2,0.2,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("x Ekseni veri ismi")
figureAxes.set_ylabel("y Ekseni veri ismi")
figureAxes.set_title("grafik bal")
```

```python
figure2 = plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])# srasyla sa sol geniilik ykseklik
eksen2=figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni byk")
eksen1.set_ylabel("Y ekseni byk")
eksen1.set_title("Ana grafik bal")


eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X ekseni kk")
eksen2.set_ylabel("Y ekseni kk")
eksen2.set_title("Kk grafik bal")

eksen2.plot(numpyDizisi2,numpyDizisi1,"r")
```

```python
plt.figure()
```

```python
(benimfigure,benimeksenler)=plt.subplots(nrows=1,ncols=2)
type(benimeksenler)

for eksen in benimeksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
plt.tight_layout()
    
#benimfigure
```

```python
benimFigure
```

```python
yenifigur=plt.figure()
#dpi=150
yeniEksen=yenifigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2, label="numpydizisi ** 2" )
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3, label="numpydizisi ** 3" )
yeniEksen.legend(loc=2)
```
