```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**2
```

```python
numpyDizisi2
```

```python
(benimfigur,benimeksen) =plt.subplots()
benimeksen.plot(numpyDizisi1,numpyDizisi2,color="#3A95A8",alpha=0.3)
benimeksen.plot(numpyDizisi2,numpyDizisi1, color="#C96F23")
```

```python
(yeniFigur,yeniEksen)=plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,color="pink",linewidth=5.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,color="black",linewidth=8.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,color="blue",linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,color="purple",linestyle="--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,color="red",linestyle=":")

yeniEksen.plot(numpyDizisi1,numpyDizisi1+7,color="blue",linestyle="--",marker="o",markersize=8,markerfacecolor="orange")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+8,color="blue",linestyle="--",marker="+",markersize=4,markerfacecolor="green")


```
