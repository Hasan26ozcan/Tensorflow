```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test =scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model=Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x_train,y_train,epochs=300,validation_data=(x_test,y_test),batch_size=250)
#validation_data=x_test ve y_test kÄ±yaslama iÅŸlemini Ã§ok daha kolay yapmamÄ±zÄ± saÄŸlÄ±yor
```

```python
kayÄ±pVerisi= pd.DataFrame(model.history.history)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
tahmindizisi=model.predict(x_test)
```
